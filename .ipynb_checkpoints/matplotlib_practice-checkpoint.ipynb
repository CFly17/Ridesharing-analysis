```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
# The plt.show() function looks for all the active figure objects 
# and opens a window that displays them 
# if you have two or more sets of data to plot.
```

```python
# 5.4 
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker=".", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# 5.5
# Bar charts -- where we can clearly see the differences among the groups and the bar color of each group.
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot
plt.barh(y_axis, x_axis)
# the bar should convey an amount on the x-axis, not a date
```

```python
# To invert the y-axis to have the months in ascending order, 
# use the gca() method. The gca() method means "get current axes." 
# We can chain the gca() method to the invert_yaxis() method 
# by using gca().invert_yaxis(), as shown here:
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
```

```python
# 5.7
# A scatter plot visually depicts a positive, negative, or no-correlation relationship between two factors.
# When we use the plt.plot() function to create a scatter plot, 
# we need to add a lowercase "o" as a parameter inside the parentheses. 
# This switches the plot from a line chart to a scatter plot.
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis)
```

```python
# To change the size of our marker on the graph, 
# we have to use the s, or size parameter, for the marker. 
# The s parameter must be equal to a feature of our data
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# We can adjust the size by multiplying the data in the y-axis by any value.
# we can iterate through the y-axis data and 
# multiply each data point by 3 and add it to a new list, like this:

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

# Then we can use the new y-axis list for the s parameter:
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#  You can use list comprehension to replace many for and while loops. 
# List comprehension is faster because it is optimized for Python 
# to spot a predictable pattern during looping.

# Syntax:
# new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Bubble chart
# To create a bubble chart using the object-oriented interface approach, 
# we'll add the s parameter just like we did when we used the MATLAB approach.

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# 5.8 
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# In general, it is a best practice to choose lighter, brighter, and analogous colors over darker tones.
# We'll add the "colors" parameter in the pie() function with our other parameters, 
# but first we need to assign a "colors" variable to an array with the names of our 12 colors.
# Assign 12 colors, one for each month.

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
# The scatter plot didn't do a great job of getting that information across, 
# but the bubble chart we created in the Skill Drill did a decent job of conveying the same 
# information as the line, bar, and pie charts, because the bubbles increased in diameter 
# as the average fare increased. Still, the bubble chart may not be a good chart to use 
# for this particular dataset, because there are only two data points for this scatter plot: 
# the months on the y-axis and the fare amount on the x-axis. 
# If we had a third parameter that could be the size of the marker, 
# like the number of riders for each month, then using the scatter plot might be a good choice.
```

```python
# 5.9 
# Adding error bars can show either the standard deviation, standard error, 
# confidence intervals, or minimum and maximum values of a dataset.

# Syntax:
# plt.errorbar(x, y, yerr=<value>)

# Import dependencies.
import matplotlib.pyplot as plt
import statistics

# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```

```python
# We can make the error bars look more professional by adding a "cap" to the error bars 
# that's equal to a scalar value. To do this, we use the capsize= parameter. 

plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# object-oriented approach:
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Object-oriented approach
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# Ticks
# To adjust the major x-axis ticks on a horizontal bar chart, we use the xticks() function.

import numpy as np
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()

#  We defined the lower and upper limit of the x-axis 
# and the increment frequency with the code: 
# np.arange(0, 51, step=5.0).
```

```python
# Object-oriented code
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
# To add minor ticks to the x-axis on a horizontal bar chart, 
# we use the set_minor_locator function with the xaxis class in the object-oriented method, like this:

# Syntax
# x.xaxis.set_minor_locator()

# Inside the set_minor_locator function, we add the MultipleLocator method 
# and pass how often we want our minor ticks.

from matplotlib.ticker import MultipleLocator

# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
